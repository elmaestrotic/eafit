**Break – Ejemplo**

```python
while(True):
  estatura = float(input("Ingrese su estatura: "))
  if(estatura > 1.34):
    print("Bienvenido a la montaña rusa")
    break
  else:
    print("No puedes ingresar a esta atracción")

print("Fin programa")
```
